```python
import pandas as pd
import numpy as np
```

```python
file_path = "Resources/COVID-19_Case_Surveillance_Public_Use_Data_with_Geography.csv"
covid_df = pd.read_csv(file_path)
covid_df.head()
```

```python
covid_df.columns
```

```python
nc_covid_df = covid_df.loc[covid_df['res_state'] == "NC"]
nc_covid_df.head()
```

```python
nc_covid_df['age_group'].unique()
```

```python
nc_covid_df = nc_covid_df[covid_df['age_group'] != 'nan']
nc_covid_df.head()
```

```python
nc_covid_df['sex'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['sex'] != 'Missing']
nc_covid_df = nc_covid_df[nc_covid_df['sex'] != 'nan']
nc_covid_df.head()
```

```python
nc_covid_df['race'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['race'] != 'Unknown']
nc_covid_df = nc_covid_df[nc_covid_df['race'] != 'nan']
nc_covid_df.head()
```

```python
nc_covid_df['ethnicity'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['ethnicity'] != 'Unknown']
nc_covid_df = nc_covid_df[nc_covid_df['ethnicity'] != 'nan']
nc_covid_df.head()
```

```python
nc_covid_df['symptom_status'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['symptom_status'] != 'Unknown']
nc_covid_df = nc_covid_df[nc_covid_df['symptom_status'] != 'Missing']
nc_covid_df.head()
```

```python
nc_covid_df['hosp_yn'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['hosp_yn'] != 'Unknown']
nc_covid_df.head()
```

```python
nc_covid_df['death_yn'].unique()
```

```python
nc_covid_df = nc_covid_df[nc_covid_df['death_yn'] != 'nan']
nc_covid_df.head()
```

```python
nc_covid_df = nc_covid_df.drop(['exposure_yn', 'underlying_conditions_yn', 'case_onset_interval'], axis=1)
nc_covid_df.head()
```

```python
nc_covid_df.dropna()
```

```python
nc_covid_df.to_csv('resources/nc_covid_data.csv', encoding = 'utf-8')
```
